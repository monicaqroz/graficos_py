```python
import matplotlib.pyplot as plt

estados = 'Amazonas', 'Rio Grande do Sul', 'Pará', 'São Paulo'
valores = [3.8, 11.2, 8, 44]
explode = (0.3, 0, 0, 0)  

cores = ["#04F504","#fd7698", "#0f5fff", "#FC5D02"]

fig1, ax1 = plt.subplots()
ax1.pie(valores, explode=explode, labels=estados, autopct='%1.1f%%', shadow=False, startangle=90, colors=cores)
ax1.axis('equal')  

plt.show()
```
